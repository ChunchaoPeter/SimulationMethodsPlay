## Demo 1: Eastimate Pi

```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
@tf.function
def monte_carlo_pi_tf(num_samples):
    """Estimate the value of π using Monte Carlo sampling with TensorFlow."""
    # Generate random x and y coordinates uniformly in [0, 1]
    x = tf.random.uniform(shape=[num_samples], minval=0, maxval=1, dtype=tf.float32)
    y = tf.random.uniform(shape=[num_samples], minval=0, maxval=1, dtype=tf.float32)
    # Determine whether points fall inside the unit quarter circle
    inside_circle = tf.less_equal(x ** 2 + y ** 2, 1.0)
    # Count the number of points inside the circle
    num_inside = tf.reduce_sum(tf.cast(inside_circle, tf.int32))
    # Estimate π (area ratio × 4)
    pi_approx = 4.0 * tf.cast(num_inside, tf.float32) / tf.cast(num_samples, tf.float32)
    
    return pi_approx, x, y, inside_circle
```

```python
# Call the function and then visualize
num_samples = 1_000_000
pi_approx, x, y, inside_circle = monte_carlo_pi_tf(num_samples)

# Now you can use .numpy() in eager mode
print(f"Monte Carlo Approximation of π: {pi_approx.numpy():.6f}")

# Visualization
plt.figure(figsize=(6, 6))
inside_mask = inside_circle.numpy()
plt.scatter(x.numpy()[inside_mask], y.numpy()[inside_mask], color='blue', s=0.05, label='Inside Circle')
plt.scatter(x.numpy()[~inside_mask], y.numpy()[~inside_mask], color='red', s=0.05, label='Outside Circle')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title(f'Monte Carlo Approximation of π: {pi_approx.numpy():.6f}')
plt.legend(loc='upper right', markerscale=10)
plt.show()
```

## Demo 2: Estimating the Area Under a Curve (Integration)

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
@tf.function
def monte_carlo_integration(num_samples, func_tensor):
    """Estimate the integral of a function using Monte Carlo sampling."""
    # Generate random x values in [0, 1]
    x = tf.random.uniform(shape=[num_samples], minval=0, maxval=1, dtype=tf.float32)
    # Generate random y values in [0, 1]
    y = tf.random.uniform(shape=[num_samples], minval=0, maxval=1, dtype=tf.float32)
    
    # Evaluate the function at x points
    f_x = func_tensor(x)
    
    # Check if points are under the curve
    under_curve = tf.less_equal(y, f_x)
    
    # Count points under the curve
    num_under = tf.reduce_sum(tf.cast(under_curve, tf.int32))
    
    # Estimate the area (ratio × total area of rectangle)
    area_approx = tf.cast(num_under, tf.float32) / tf.cast(num_samples, tf.float32)
    
    return area_approx, x, y, under_curve, f_x

# Define the function to integrate (e.g., x^2)
@tf.function
def func(x):
    return x ** 2
```

```python
# Run the Monte Carlo simulation
num_samples = 50000
area_approx, x, y, under_curve, f_x = monte_carlo_integration(num_samples, func)

print(f"Monte Carlo approximation of ∫x² dx from 0 to 1: {area_approx.numpy():.6f}")
print(f"Actual value: {1/3:.6f}")

# Visualization
plt.figure(figsize=(8, 6))
under_mask = under_curve.numpy()
plt.scatter(x.numpy()[under_mask], y.numpy()[under_mask], color='green', s=1, alpha=0.5, label='Under Curve')
plt.scatter(x.numpy()[~under_mask], y.numpy()[~under_mask], color='red', s=1, alpha=0.5, label='Above Curve')

# Plot the actual function
x_line = np.linspace(0, 1, 100)
plt.plot(x_line, x_line**2, 'blue', linewidth=2, label='y = x²')

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Monte Carlo Integration: Area ≈ {area_approx.numpy():.6f} (Actual: 0.333)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
