```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ============================================================================
# METROPOLIS-HASTINGS ALGORITHM
# ============================================================================
```

```python


def metropolis_hastings_tf(logp_fn, x_init=0.0, n_samples=20_000, 
                          proposal_std=1.0, burn_in=2_000):
    """
    Metropolis-Hastings MCMC with symmetric random walk proposal.
    
    Parameters:
    -----------
    logp_fn : callable
        Log of unnormalized target density log f(x)
    x_init : float
        Initial state X^(0)
    n_samples : int
        Total iterations (including burn-in)
    proposal_std : float
        Standard deviation of proposal N(x_t, proposal_std²)
    burn_in : int
        Number of initial samples to discard
    
    Returns:
    --------
    samples : tensor
        Samples after burn-in
    acceptance_rate : float
        Fraction of accepted proposals
    chain : tensor
        Full chain including burn-in
    
    Algorithm:
    ----------
    1. Propose: X' ~ N(X_t, proposal_std²)
    2. Compute: α = min(1, f(X')/f(X_t))
       In log space: log α = min(0, log f(X') - log f(X_t))
    3. Accept with probability α
    """
    
    # Initialize
    chain = tf.TensorArray(dtype=tf.float32, size=n_samples, dynamic_size=False)
    x_current = tf.constant(x_init, dtype=tf.float32)
    logp_current = logp_fn(x_current)
    n_accepted = tf.constant(0, dtype=tf.int32)
    
    # Main MCMC loop
    for t in tf.range(n_samples):
        # Step 1: Propose new state
        x_proposal = x_current + tf.random.normal([], mean=0.0, stddev=proposal_std)
        logp_proposal = logp_fn(x_proposal)
        
        # Step 2: Compute log acceptance probability
        # For symmetric proposal: log α = log f(x') - log f(x)
        log_alpha = logp_proposal - logp_current
        
        # Step 3: Accept or reject (log-space for numerical stability)
        log_u = tf.math.log(tf.random.uniform([]))
        
        if log_u < log_alpha:
            x_current = x_proposal
            logp_current = logp_proposal
            n_accepted = n_accepted + 1
        
        # Store current state
        chain = chain.write(t, x_current)
    
    # Convert to tensor
    chain = chain.stack()
    acceptance_rate = tf.cast(n_accepted, tf.float32) / tf.cast(n_samples, tf.float32)
    
    # Remove burn-in
    samples = chain[burn_in:]
    
    return samples, acceptance_rate.numpy(), chain


```

```python
def unnormalized_logp(x):
    """
    Log of unnormalized target: log f(x) = -x^4 + 2x^2
    This is a bimodal distribution with peaks around x ≈ ±1
    """
    return -x**4 + 2.0 * x**2


def true_density(x):
    """Compute normalized target density for comparison"""
    unnorm = np.exp(unnormalized_logp(x))
    x_grid = np.linspace(-3, 3, 10000)
    Z = np.trapz(np.exp(unnormalized_logp(x_grid)), x_grid)
    return unnorm / Z
```

```python
def run_example():
    """Run M-H on bimodal distribution with minimal plots"""
    
    print("Running Metropolis-Hastings on Bimodal Distribution")
    print("=" * 60)
    
    # Run MCMC
    samples, acceptance_rate, chain = metropolis_hastings_tf(
        logp_fn=unnormalized_logp,
        x_init=0.0,
        n_samples=20_000,
        proposal_std=1.0,
        burn_in=2_000
    )
    
    # Print statistics
    samples_np = samples.numpy()
    print(f"\nAcceptance rate: {acceptance_rate:.2%}")
    print(f"Samples (after burn-in): {len(samples_np)}")
    print(f"Sample mean: {np.mean(samples_np):.4f}")
    print(f"Sample std: {np.std(samples_np):.4f}")
    
    # Create 2 essential plots only
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    
    # Plot 1: Trace plot
    axes[0].plot(chain.numpy(), alpha=0.7, linewidth=0.5)
    axes[0].axvline(x=2000, color='red', linestyle='--', label='Burn-in end')
    axes[0].set_xlabel('Iteration')
    axes[0].set_ylabel('x')
    axes[0].set_title('Trace Plot')
    axes[0].legend()
    axes[0].grid(alpha=0.3)
    
    # Plot 2: Histogram vs True Density
    x_grid = np.linspace(-2.5, 2.5, 1000)
    true_dens = true_density(x_grid)
    
    axes[1].hist(samples_np, bins=50, density=True, alpha=0.6, 
                 color='skyblue', edgecolor='black', label='MCMC')
    axes[1].plot(x_grid, true_dens, 'r-', linewidth=2, label='True density')
    axes[1].set_xlabel('x')
    axes[1].set_ylabel('Density')
    axes[1].set_title(f'Distribution (Acceptance: {acceptance_rate:.1%})')
    axes[1].legend()
    axes[1].grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return samples, acceptance_rate, chain
```

```python
# ============================================================================
# RUN EXAMPLE
# ============================================================================
if __name__ == "__main__":
    samples, acc_rate, chain = run_example()
    print("\nDone!")
```
